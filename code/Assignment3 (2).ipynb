In this [paper](https://www.microsoft.com/en-us/research/wp-content/uploads/2016/02/Ferreira_Fisher_Sample_Oriented_Tasks.pdf) the authors describe the challenges users face when trying to make judgements about probabilistic data generated through samples. As an example, they look at a bar chart of four years of data (replicated below in Figure 1). Each year has a y-axis value, which is derived from a sample of a larger dataset. For instance, the first value might be the number votes in a given district or riding for 1992, with the average being around 33,000. On top of this is plotted the 95% confidence interval for the mean (see the boxplot lectures for more information, and the yerr parameter of barcharts).

A challenge that users face is that, for a given y-axis value (e.g. 42,000), it is difficult to know which x-axis values are most likely to be representative, because the confidence levels overlap and their distributions are different (the lengths of the confidence interval bars are unequal). One of the solutions the authors propose for this problem (Figure 2c) is to allow users to indicate the y-axis value of interest (e.g. 42,000) and then draw a horizontal line and color bars based on this value. So bars might be colored red if they are definitely above this value (given the confidence interval), blue if they are definitely below this value, or white if they contain this value.

```python
# Use the following data for this assignment:

import pandas as pd
import numpy as np

np.random.seed(12345)

df = pd.DataFrame([np.random.normal(32000,200000,3650), 
                   np.random.normal(43000,100000,3650), 
                   np.random.normal(43500,140000,3650), 
                   np.random.normal(48000,70000,3650)], 
                  index=[1992,1993,1994,1995])
df
```

```python
from matplotlib import pyplot as plt
import scipy.stats as st

##Creating stats dataframe to get the mean and standard deviation
stat_check = df.transpose()
stat_check = pd.DataFrame(stat_check.describe())
stat_check
```

```python
##Creates columns for the standard error, mean, and stardard deviation
df['err'] = df.transpose().sem()
df['mean'] = stat_check.iloc[[1]].transpose()
df['std'] = stat_check.iloc[[2]].transpose()

##reset to pull year values
df.reset_index(inplace=True)
X_val = df['index']

## using st.norm.interval() to find the 95% confidence interval
CI = st.norm.interval(alpha=0.95, loc=df['mean'], scale=df['err'])

##convert the array to a list
CI = np.array(CI).tolist()

##convert the list to a dataframe
CI = pd.DataFrame(CI, columns = ['1992','1993','1994','1995']).transpose().reset_index()

##Add confidence interval dataframe elements to the main dataframe
CI.columns = ["index","conf_low","conf_high"]
df['conf_low'] = CI['conf_low']
df['conf_high'] = CI['conf_high']
df['conf_diff'] = df['conf_high'] - df['conf_low']

##make the years the index again
df.set_index('index', inplace=True)
```

```python
df
```

```python
#import matplotlib.lines as mlines
%matplotlib notebook
import matplotlib.pyplot as plt
from matplotlib import cm
import matplotlib.ticker as ticker

##Coloring for the plot
max_color='red'
min_color='blue'
neutral_color='grey'

class Cursor(object):
    _df=None
    _bl=None
    def __init__(self, ax, data_F, bars):
        self._df=data_F
        self._bl=bars
        self.ax = ax
        self.lx = ax.axhline(color='b')

    def mouse_move(self, event):
        if not event.inaxes:
            return
        x, y = event.xdata, event.ydata
        self.lx.set_ydata(y)
        for index, row in self._df.iterrows():
            ##Will change the color to grey if the cursor y value if above the higher confidence interval
            if row['conf_high']<y:
                self._bl[self._df.index.get_loc(index)].set_color(neutral_color)
                continue
            ##Will change the color to red if the cursor y value if above the lower confidence interval
            if row['conf_low']>y:
                self._bl[self._df.index.get_loc(index)].set_color(max_color)
                continue
            
            self._bl[self._df.index.get_loc(index)].set_color(min_color)
        plt.draw()
        
def plotting(x_title , y_title, fig_title, c_alpha=0.5):
    ##Creating the main plot
    ax=df['mean'].plot.bar(yerr=df['err'], 
        title =fig_title, figsize=(5, 5), 
        legend=False, fontsize=10, alpha=c_alpha, width=0.95, 
        rot=0, position=0, style='-', color=neutral_color)
    ax.set_xlabel(x_title)
    ax.set_ylabel(y_title)
    #Remove spines for top and right of plot
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)

    ##Setting location of the tick marks so they are in the center of the bars
    ax.xaxis.set_major_locator(ticker.FixedLocator([0.5,1.5,2.5,3.5]))
    
    return (ax, ax.get_children()[1:5])

##running the function for the plot
ax, bar=plotting("Year", "Random Normal Mean", "Easiest option")

## Passes plot information to the cursor function
cursor = Cursor(ax, df, bar)
plt.connect('motion_notify_event', cursor.mouse_move)

```
